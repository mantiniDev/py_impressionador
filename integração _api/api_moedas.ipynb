```python
import requests
import json

cotacoes = requests.get('https://economia.awesomeapi.com.br/all')
cotacao_dic = cotacoes.json()
print(f'Testando {cotacao_dic}')
```

```python
print(f'A cotação do Dólar é {cotacao_dic["USD"]["bid"]}')
print(f'A cotação do Euro é {cotacao_dic["EUR"]["bid"]}')
print(f'A cotação do dólar é {cotacao_dic["BTC"]["bid"]}')
```

```python
dolar_30d = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/30')
dolar_30_dic = dolar_30d.json()
lista_cotacoes_30d = [item['bid'] for item in dolar_30_dic]
print(lista_cotacoes_30d)

#print(dolar_30_dic[0])
#print(f'A cotação do dólar é {dolar_30_dic["USD"]["bid"]}')
```

```python
cotacoes_btc = requests.get('https://economia.awesomeapi.com.br/BTC-BRl/200?start_date=20200101&end_date=20201031')

btc_dic = cotacoes_btc.json()
#print(btc_dic)
lista_cotacoes_btc = [item['bid'] for item in btc_dic]
lista_cotacoes_btc.reverse()
print(lista_cotacoes_btc)
print(len(lista_cotacoes_btc))
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(15,5))
plt.plot(lista_cotacoes_btc)
plt.show()
```
